```python
# set filename
filename = "https://raw.githubusercontent.com/vaibhavwalvekar/NYC-Flights-2013-Dataset-Analysis/master/flights.csv"
```

```python
import pandas as pd

pd_file = pd.read_csv(filename, index_col=0)
```

```python
pd_file.head()
```

```python
import dask.dataframe as dd

df = dd.from_pandas(pd_file, npartitions=2)
df.head()
```

```python
df.describe()
```

```python
df.dep_delay.max(axis=0).compute()
```

```python
%%time

df.groupby("carrier").dep_delay.max().compute()
```
